```python
# Initialize the Hugging Face API
api = HfApi()

# Fetch the list of models
models = api.list_models()

# Print details of the models
count = 0
for model in models:
    count = count + 1
    print(f"Model ID: {model.modelId}")
    print(f"Model Name: {model.modelId}")
    if count == 100:
        break
```
