```python
from gtts import gTTS
from IPython.display import Audio

text = """
Hello. This is a simple Python program.
It converts English text into English speech.
The audio file will be saved in Google Colab.
"""

tts = gTTS(text=text, lang="en")

file_name = "english_speech.mp3"
tts.save(file_name)

print("Audio file saved as:", file_name)

Audio(file_name)
```

```python
from google.colab import files

files.download("english_speech.mp3")
```
