```python
import os
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_google_genai import ChatGoogleGenerativeAI

#os.environ["GOOGLE_API_KEY"] = "YOUR_GEMINI_API_KEY"
```

```python
@tool
def save_blog_to_file(blog_text: str) -> str:
    """Save the final blog article to a markdown file."""
    file_name = "generated_blog.md"

    with open(file_name, "w", encoding="utf-8") as f:
        f.write(blog_text)

    return f"Blog saved successfully as {file_name}"


model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.7
)

agent = create_agent(
    model=model,
    tools=[save_blog_to_file],
    system_prompt="""
    You are a professional blog writing agent.

    Your task:
    1. Understand the blog topic.
    2. Create a clear blog title.
    3. Write an introduction.
    4. Write 4 to 6 useful sections.
    5. Add examples where needed.
    6. Write a conclusion.
    7. Save the final blog using the save_blog_to_file tool.

    Blog style:
    - Simple English
    - Beginner friendly
    - Practical and useful
    - Use Markdown formatting
    """
)

topic = input("Enter blog topic: ")

response = agent.invoke({
    "messages": [
        {
            "role": "user",
            "content": f"Write a complete blog article on: {topic}"
        }
    ]
})

print(response["messages"][-1].content)
```
