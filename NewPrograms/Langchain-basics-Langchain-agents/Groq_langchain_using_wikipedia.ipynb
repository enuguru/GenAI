```python
import os
import wikipedia

from langchain_groq import ChatGroq

#os.environ["GROQ_API_KEY"] = "YOUR_GROQ_API_KEY"

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0
)

question = "Who was Nikola Tesla?"

# Search Wikipedia
search_results = wikipedia.search(question)

if not search_results:
    print("No Wikipedia result found.")
else:
    page_title = search_results[0]
    print("Wikipedia page found:", page_title)

    summary = wikipedia.summary(page_title, sentences=5)

    prompt = f"""
    Answer the user's question using only the Wikipedia text below.

    User question:
    {question}

    Wikipedia text:
    {summary}

    Answer in simple English.
    """

    response = llm.invoke(prompt)

    print("\nAnswer:\n")
    print(response.content)
```
