```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.utilities import WikipediaAPIWrapper
from langchain_community.tools import WikipediaQueryRun
from langchain_core.prompts import ChatPromptTemplate
```

```python
import os
from google.colab import userdata
from google import genai
# Load API key from Colab Secrets into environment variable
os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")
```

```python
# Set Gemini API Key
# os.environ["GOOGLE_API_KEY"] = "your_google_api_key"

# Create Gemini LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)

# Wikipedia tool
wiki_tool = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())

# -------- Step 1: Retrieve Information --------

def retrieve_information(topic):
    return wiki_tool.run(topic)

# -------- Step 2: Summarize --------

summary_prompt = ChatPromptTemplate.from_template(
"""
Summarize the following information in 5 concise sentences.

Information:
{context}
"""
)

def summarize(context):
    prompt = summary_prompt.format(context=context)
    response = llm.invoke(prompt)
    return response.content

# -------- Step 3: Generate Quiz --------

quiz_prompt = ChatPromptTemplate.from_template(
"""
Based on the summary below, generate 3 quiz questions.

Summary:
{summary}
"""
)

def generate_quiz(summary):
    prompt = quiz_prompt.format(summary=summary)
    response = llm.invoke(prompt)
    return response.content

```

```python
# -------- Chain Execution --------

topic = input("Enter a topic: ")

print("\nStep 1: Retrieving Wikipedia information...")
context = retrieve_information(topic)

print("\nStep 2: Generating summary...")
summary = summarize(context)
print("\nSummary:\n", summary)

print("\nStep 3: Generating quiz questions...")
quiz = generate_quiz(summary)

print("\nQuiz Questions:\n", quiz)
```
