```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
# from langchain.schema import Document
from langchain_core.documents import Document
```

```python
# Step 1: Create embedding model
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Step 2: Create documents
documents = [
    Document(page_content="Nikola Tesla was a Serbian-American inventor and electrical engineer."),
    Document(page_content="Albert Einstein developed the theory of relativity."),
    Document(page_content="Python is a popular programming language used for AI and data science."),
]

# Step 3: Store documents in FAISS vector database
vector_db = FAISS.from_documents(documents, embedding_model)

# Step 4: Query for semantic search
query = input("Enter your query: ")

results = vector_db.similarity_search(query)

# Step 5: Show results
print("\nMost relevant document:")
print(results[0].page_content)
```
