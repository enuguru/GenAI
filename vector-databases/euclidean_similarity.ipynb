```python
import spacy
import numpy as np

# Load the spacy model for word embeddings
nlp = spacy.load("en_core_web_md")

def get_word_vector(word):
    """Get the vector for a word."""
    return nlp(word).vector

def euclidean_distance(vec1, vec2):
    """Compute the Euclidean distance between two vectors."""
    return np.linalg.norm(vec1 - vec2)

def euclidean_similarity(vec1, vec2):
    """Convert Euclidean distance to a similarity measure."""
    distance = euclidean_distance(vec1, vec2)
    print(distance)
    # Transform distance to similarity (1 / (1 + distance) for stability)
    similarity = 1 / (1 + distance)
    return similarity

# Input words
word1 = "king"
word2 = "queen"

# Get embeddings
vector1 = get_word_vector(word1)
vector2 = get_word_vector(word2)

# Calculate Euclidean similarity
similarity = euclidean_similarity(vector1, vector2)

print(f"Euclidean Similarity between '{word1}' and '{word2}': {similarity:.4f}")
```
