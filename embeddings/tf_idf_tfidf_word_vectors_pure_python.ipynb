```python
import math
import re

# --------------------------------
# Documents
# --------------------------------

documents = [
    "Machine learning is useful",
    "Deep learning is part of machine learning",
    "Artificial intelligence uses machine learning",
    "Python is useful for artificial intelligence"
]

# --------------------------------
# Tokenization
# --------------------------------

def tokenize(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", "", text)
    return text.split()

tokenized_docs = [tokenize(doc) for doc in documents]

# --------------------------------
# Vocabulary
# --------------------------------

vocabulary = sorted(
    set(word for doc in tokenized_docs for word in doc)
)

print("=" * 80)
print("VOCABULARY")
print("=" * 80)
print(vocabulary)

# --------------------------------
# TF
# --------------------------------

def compute_tf(tokens, vocabulary):

    total_words = len(tokens)

    tf = {}

    for word in vocabulary:
        tf[word] = tokens.count(word) / total_words

    return tf

tf_scores = [
    compute_tf(doc, vocabulary)
    for doc in tokenized_docs
]

# --------------------------------
# IDF
# --------------------------------

def compute_idf(tokenized_docs, vocabulary):

    N = len(tokenized_docs)

    idf = {}

    for word in vocabulary:

        doc_count = 0

        for doc in tokenized_docs:
            if word in doc:
                doc_count += 1

        idf[word] = math.log(N / doc_count)

    return idf

idf_scores = compute_idf(
    tokenized_docs,
    vocabulary
)

# --------------------------------
# TF-IDF
# --------------------------------

def compute_tfidf(tf, idf, vocabulary):

    tfidf = {}

    for word in vocabulary:
        tfidf[word] = tf[word] * idf[word]

    return tfidf

tfidf_scores = [
    compute_tfidf(tf, idf_scores, vocabulary)
    for tf in tf_scores
]

# ==========================================
# PRINT TF FOR ALL DOCUMENTS
# ==========================================

print("\n")
print("=" * 80)
print("TERM FREQUENCY (TF)")
print("=" * 80)

for i, tf in enumerate(tf_scores):

    print(f"\nDocument {i+1}")
    print(documents[i])

    for word in vocabulary:
        print(
            f"{word:15s} : {tf[word]:.4f}"
        )

# ==========================================
# PRINT IDF
# ==========================================

print("\n")
print("=" * 80)
print("INVERSE DOCUMENT FREQUENCY (IDF)")
print("=" * 80)

for word in vocabulary:

    print(
        f"{word:15s} : {idf_scores[word]:.4f}"
    )

# ==========================================
# PRINT TF-IDF FOR ALL DOCUMENTS
# ==========================================

print("\n")
print("=" * 80)
print("TF-IDF")
print("=" * 80)

for i, tfidf in enumerate(tfidf_scores):

    print(f"\nDocument {i+1}")
    print(documents[i])

    for word in vocabulary:

        print(
            f"{word:15s} : {tfidf[word]:.4f}"
        )

# ==========================================
# PRINT NUMERIC TF-IDF VECTORS
# ==========================================

print("\n")
print("=" * 80)
print("TF-IDF NUMERIC VECTORS")
print("=" * 80)

for i, tfidf in enumerate(tfidf_scores):

    vector = [
        tfidf[word]
        for word in vocabulary
    ]

    print(f"\nDocument {i+1}")
    print(vector)
```
