```python
import os

os.environ["OPENAI_API_KEY"] = "" # Replace with your OpenAI API key
```

```python
from crewai import Agent, Task, Crew

# -----------------------------
#  Define Agents
# -----------------------------

researcher = Agent(
    name="Researcher",
    role="AI Research Specialist",
    goal="Search and extract relevant insights about the given topic",
    backstory=(
        "You are an expert AI researcher. You find accurate, reliable information"
        " and explain it in simple terms."
    )
)
```

```python
writer = Agent(
    name="Writer",
    role="Content Writer",
    goal="Write a clear and concise summary of the topic based on research",
    backstory=(
        "You are a skilled writer who produces engaging and easy-to-understand content."
    )
)
```

```python
# -----------------------------
#  Define Tasks
# -----------------------------

research_task = Task(
    description="Research the topic: 'Agentic AI systems and their benefits'",
    agent=researcher,
    expected_output="A list of key insights about agentic AI"
)

write_task = Task(
    description="Using the researcher's insights, write a 3-paragraph summary.",
    agent=writer,
    expected_output="A well-written summary explaining agentic AI systems"
)
```

```python
# -----------------------------
#  Create the Crew
# -----------------------------

crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
)
```

```python
# -----------------------------
#  Run the Crew
# -----------------------------

result = crew.kickoff()
print("\n\n=== FINAL OUTPUT ===\n")
print(result)
```
