```python
from crewai import Agent, Task, Crew
import os
```

```python
from crewai import LLM
```

```python
from google.colab import userdata
# --------------------------------
# SET GEMINI API KEY
# --------------------------------
os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")
#YOUR_GOOGLE_API_KEY = "GOOGLE_API_KEY"
```

```python
# --------------------------------
# GEMINI 2.5 FLASH MODEL
# --------------------------------
from langchain_google_genai import ChatGoogleGenerativeAI

#llm = ChatGoogleGenerativeAI(
#    model="gemini-2.5-flash",
#    temperature=0.3
#)

llm = LLM(
    model="gemini/gemini-2.5-flash",
    temperature=0.3,
    #api_key=userdata.get("GOOGLE_API_KEY")
)
```

```python
# --------------------------------
# AGENTS
# --------------------------------

idea_generator = Agent(
    role="Startup Idea Generator",
    goal="Generate innovative AI startup ideas",
    backstory="You are a creative AI entrepreneur who spots business opportunities.",
    llm=llm,
    verbose=True
)

market_analyst = Agent(
    role="Market Analyst",
    goal="Evaluate startup ideas for feasibility and market demand",
    backstory="You analyze business ideas and assess their potential.",
    llm=llm,
    verbose=True
)

business_planner = Agent(
    role="Business Planner",
    goal="Create a simple execution plan for promising ideas",
    backstory="You convert ideas into practical action plans.",
    llm=llm,
    verbose=True
)
```

```python


# --------------------------------
# TASKS
# --------------------------------

idea_task = Task(
    description="Generate one innovative AI startup idea in the healthcare domain.",
    expected_output="A clear startup idea with problem statement and proposed AI solution.",
    agent=idea_generator
)

analysis_task = Task(
    description="Evaluate the proposed startup idea for market demand, competition, and feasibility.",
    expected_output="A structured evaluation including strengths, risks, and opportunity score.",
    agent=market_analyst
)

plan_task = Task(
    description="Create a simple 5-step execution roadmap for launching this startup.",
    expected_output="A clear step-by-step execution plan.",
    agent=business_planner
)
```

```python
from crewai import Process
```

```python


# --------------------------------
# CREW
# --------------------------------

crew = Crew(
    agents=[idea_generator, market_analyst, business_planner],
    tasks=[idea_task, analysis_task, plan_task],
    process=Process.sequential,   # Important for teaching
    verbose=True
)
```

```python

# --------------------------------
# EXECUTE
# --------------------------------

result = crew.kickoff()

print("\n\n===== FINAL OUTPUT =====\n")
print(result)
```
